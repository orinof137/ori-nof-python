### Part 2: Checking Variable Types
Use the type() function to check what type a variable is:  
Code cell:

```python
# Let's check different types
age = 15
height = 1.75
name = "David"
is_student = True

print(type(age))           # <class 'int'>
print(type(height))        # <class 'float'>
print(type(name))          # <class 'str'>
print(type(is_student))    # <class 'bool'>
```

---
### Part 3: Type Conversion
Sometimes we need to convert between types:  
Converting to Integer

```python
text = "25"
number = int(text)
print(number + 5)  # Result: 30
```

Converting to Float


```python
text = "3.14"
number = float(text)
print(number * 2)  # Result: 6.28
```

Converting to String


```python
age = 15
text = str(age)
print("I am " + text + " years old")
```

Important! You can't convert text that isn't a number:


```python
int("hello")  # This will cause an ERROR!
```

### Part 4: Working with User Input
The input() function lets us get information from the user:

```python
name = input("What is your name? ")
print(f"Hello {name}!")
```

Important: input() always returns a string, even if the user types a number!

```python
# Wrong way:
age = input("How old are you? ")
future = age + 10  # ERROR! Can't add number to string

# Correct way:
age = int(input("How old are you? "))
future = age + 10  # This works!
```
