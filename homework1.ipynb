```python
# Tip Calculator

bill = float(input("Enter your bill amount: $"))
tip_percent = int(input("Enter tip percentage (10, 15, or 20): "))

tip_amount = bill * (tip_percent / 100)
total = bill + tip_amount

print(f"Tip amount: ${tip_amount:.2f}")
print(f"Total bill (with tip): ${total:.2f}")
```

```python
# Temperature Converter: Celsius to Fahrenheit

celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = celsius * 9/5 + 32

print(f"{celsius}°C is equal to {fahrenheit:.2f}°F")
```

```python
# How Many Seconds in Your Life

age = int(input("Enter your age in years: "))

seconds_in_life = age * 365 * 24 * 60 * 60

print(f"You have lived approximately {seconds_in_life:,} seconds!")
```
